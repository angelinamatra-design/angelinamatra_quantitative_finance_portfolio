```python
#print s&p
import yfinance as yf 
import pandas as pd

spy = yf.download("SPY", start="2020-01-01", end="2026-05-01").round(2)

print(spy) 
```

```python
#describe S&P chart 
spy.describe().round(2) 
```

```python
#calculate & describe daily returns 
spy["Daily Return"] = spy["Close"].pct_change()
spy["Daily Return"].describe() 
```

```python
#add year column 
spy["Year"] = spy.index.year
print(spy["Year"])
```
